## Import and preprocess data

```
import re

with open('lyrics.txt') as f:
  data = f.read()
  # remove punctuations
  data = re.sub(r'[^\w\s]','',data) 
  # remove multiple whitespaces,tabs, newlines
  data = re.sub('\s+',' ', data)
  # convert to lowercase
  data = data.lower()
  print(data)
```

## basic WordCloud

```
from wordcloud import WordCloud
import matplotlib.pyplot as plt

wordcloud = WordCloud(max_words=50, background_color="white").generate(data)

plt.figure(figsize=(15,15))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.savefig('cloud.png')
plt.show()
```

## WordCloud with tear mask

```
# convert white pixels from 0 to 255
import numpy as np
from PIL import Image

tear_mask = np.array(Image.open("tear.png"))
invert = lambda x: 255 if x == 0 else x
tear_mask = np.vectorize(invert)(tear_mask)

tear_mask
```

```
wordcloud = WordCloud(background_color="white", max_words=50, mask=tear_mask,
        contour_width=3, contour_color='blue').generate(data)

plt.figure(figsize=(15,15))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.savefig('cloud.png')
plt.show()
```
